```python
%matplotlib ipympl
import numpy as np
import os 
import matplotlib.pyplot as plt
import re
```

```python
## paths ##
directory = '/home/tomh/Documents/projects/czebra/archived_models/230920/f1_221103_spot-label_second-half-stims_time-since-last_tectal_10000rois_2000iters_128hidden_0.001L2_0.8train/script'
filename_list = ['v5_RMSE.txt', 'v4_predict-on-train_RMSE.txt']
filename = filename_list[1]
filepath = directory + os.sep + filename
```

```python
## read in from file ##
with open(filepath, 'r') as f:
    contents = f.read()


pattern = re.compile('\d\d\d\.\d*')
values = re.findall(pattern, contents)

# convert to float
for i in range(len(values)):
    val = values[i]
    values[i] = float(val)

# convert to array
values = np.array(values)
```

```python
## plot data ##

# create x values
x = np.arange(100,2001,100)

# plot
fig, ax = plt.subplots()
ax.plot(x, values)
plt.xlabel('Training iterations')
plt.ylabel('root MSE')
plt.show()
```
